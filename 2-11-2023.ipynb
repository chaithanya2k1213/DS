```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.displot(data=df['total_bill'],kde=False,bins=5)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.rugplot(df['total_bill'])
```

```python
sns.rugplot(x='sex',y='total_bill',data=df)
```

```python
 sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.countplot(x='sex',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=df)
```

```python
sns.stripplot(x='day',y='total_bill',data=df,jitter=True)
```
